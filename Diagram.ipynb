```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots(figsize=(12, 8))

# Create a box style
box_style = dict(boxstyle="round,pad=0.3", edgecolor="black", facecolor="lightblue")

# Draw the boxes and arrows
def draw_box(ax, text, xy, box_style):
    bbox = patches.FancyBboxPatch(
        (xy[0] - 0.1, xy[1] - 0.05), 0.2, 0.1, **box_style
    )
    ax.add_patch(bbox)
    ax.text(xy[0], xy[1], text, ha="center", va="center", fontsize=12)

# Draw the process flow
steps = [
    "Input Image",
    "Preprocessing",
    "Encoder (Vision Transformer)",
    "Decoder (Text Transformer)",
    "Postprocessing",
    "Output Text"
]
positions = [
    (0.1, 0.9),
    (0.3, 0.9),
    (0.5, 0.9),
    (0.7, 0.9),
    (0.9, 0.9),
    (1.1, 0.9)
]

for i, step in enumerate(steps):
    draw_box(ax, step, positions[i], box_style)
    if i < len(steps) - 1:
        ax.arrow(
            positions[i][0] + 0.1, positions[i][1],
            positions[i+1][0] - positions[i][0] - 0.2, 0,
            head_width=0.03, head_length=0.05, fc="black", ec="black"
        )

# Add detail texts
details = [
    "Image file (e.g., JPEG, PNG)",
    "Resize, Normalize, Augmentations",
    "DeiT: Patch Embeddings, Self-Attention\nCreates feature vectors",
    "TrOCR: Self-Attention, Encoder-Decoder Attention\nGenerates text tokens",
    "Remove special tokens, Format text",
    "Final text output"
]
detail_positions = [
    (0.1, 0.75),
    (0.3, 0.75),
    (0.5, 0.75),
    (0.7, 0.75),
    (0.9, 0.75),
    (1.1, 0.75)
]

for i, detail in enumerate(details):
    ax.text(detail_positions[i][0], detail_positions[i][1], detail, ha="center", va="center", fontsize=10, wrap=True)

ax.set_xlim(0, 1.2)
ax.set_ylim(0.6, 1)
ax.axis("off")

plt.show()
```
